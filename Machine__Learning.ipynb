### **Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.**

 **Answer**   1: Simple Linear Regression is a method to find a straight-line relationship between one independend variable (input, X) amd one dependend variable (output, Y). It tries to answer "if X changes, hoe does Y change?"

For example, you might want to predict a person's weight(Y) based on their height(X). Ther purpose is to predict or understanding how one variable influnces nother useing a straight-line formula.

### **Question 2: What are the key assumptions of Simple Linear Regression?** ###

**Answer 2:** There are 4 main assumptions:

1. Linearity: The relationship between X and Y should be straight-line-like.
2. Independence: Data points must be independence of each other.
3. Homescedasticity: The spread (variance) of error should be constant across all X values.
4. Normality of Error: The residuals (prediction error) should roughly follow a normal curve.

If these assumption break badly, your regressuon becomes unreliable.

### **Question 7: Name and briefly describe three common evaluation metrics for regression models.** ###


**Answer 7:** Three commen metrics used to evaluate how well a regression model's prediction match the actual values year.

1.*Mean Absolute Error (MEA)*:

   . Desription: The average of the absolute differnce between the predicted values and the actual.

   . Interpretation: It is a measure of the average magnitudde of the error in the set of predictiond, without considering their diirection. It is in the same unit as the dependent variable.

2.*Mean Squared Error(MSE)*:

   . Desription: The average of the squared difference between the predicted and actual vslues.

   . Interpretation: Because error are squared, it heavily penalizes large error, making it useful when significant error are particularly umderiable. The units are the square of dependent variable's units.

3.*Root Mean Squared Error(RMSE)*:

   . Desription: the squred root of the MSE.
   
   . Interpretation: It is often preferred over MSE because it coverts the units back into the origunal units of the dependent variable, making it more interoreable than MSE. It still penlizes larger error more than MAE.

### **Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept. (Include your Python code and output in the code box below.)** ###

````
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

#sample data
x = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,5,4,5])

# splitting data
x_train, x_text, y_train, y_text = train_test_split(x,y, test_size=0.2, random_state=0)

# model creation and training
model = LinearRegression()
model.fit(x_train, y_train)

# output
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
````
